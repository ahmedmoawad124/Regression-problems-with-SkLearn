```
''' ***  Linear regression ( Single feature ) *** '''
import numpy as  np
X_train = np.array([[1, 2, 4, 0]]).reshape(-1,1)
Y_train = np.array([[0.5, 1, 2, 0]]).reshape(-1,1)
X_test = np.array([[0.5, 1.5, 3, 5]]).reshape(-1,1)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)

# To predict the test set:
y_predicted = regressor.predict(X_test)
print("y predicred for X testing set  = \n", y_predicted)

# Vesualization:
import matplotlib.pyplot as plt
y_from_hypothis = regressor.predict(X_train)
print("y from hypo. for X training set = \n", y_from_hypothis)


plt.scatter(X_train, Y_train, color='red')
plt.plot(X_train, y_from_hypothis, color='blue')
plt.show()
```

```
''' ***  Linear regression ( Multiple features ) *** '''
import numpy as  np
X_train = np.array([[2.75, 5.3], [2.5, 5.3], [2.25 ,5.5 ], [2.5 ,5.6 ]])
Y_train = np.array([[1464, 1394, 1195, 1254]]).reshape(-1,1)
X_test = np.array([[2, 5.7], [2, 5.7], [1.75,5.9], [1.75, 6.1]])
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)

# To predict the test set:
y_predicted = regressor.predict(X_test)
print("y predicred for X testing set  =\n", y_predicted)
print("-" * 60)
# Vesualization:
import matplotlib.pyplot as plt
y_from_hypothis = regressor.predict(X_train)
print("y from hypo. for X training set =\n", y_from_hypothis)
print("The difference between original Y and Y from the hypothis =\n", Y_train - y_from_hypothis)
m = np.size(X_train, 0)
print( " # available data in our data set :", m)
Cost_function = np.sum((Y_train - y_from_hypothis)**2) / (2*m)
print("The cost function = ",Cost_function)
```
